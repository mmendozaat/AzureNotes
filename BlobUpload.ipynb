```python
import os
os.listdir(dirs[0])
```

```python
from azure.storage.blob import ContainerClient, BlobClient

def upload(filename, dest):
    cc = ContainerClient.from_connection_string(
               'DefaultEndpointsProtocol=https;AccountName=<storageaccount>;AccountKey=<storageaccountkey>;EndpointSuffix=core.windows.net',
               'ciadatalake-eastus-final-01')
    with open(filename, 'rb') as data:
        cc.upload_blob(dest, data)
```

```python
"""
Factusage_part_0_2020_08_25_15_01_30.gz.parquet
8/25/2020, 2:36:22 PM

Factusage_part_0_2020_08_25_18_21_30.gz.parquet
8/25/2020, 2:41:06 PM
"""
import time
destpath = 'Facts/UsageLogs/FactUsage/2020/08/25'
for h in dirs:
    for m in os.listdir(h):
        hm = os.path.join(h, m)
        for s in os.listdir(hm):
            hms = os.path.join(hm, s)
            for f in os.listdir(hms):
                filename = os.path.join(hms, f)
                dest = os.path.join(destpath, hms, f.replace('.gz', '_2.gz'))
                print(filename, '->', dest)
                upload(filename, dest)
                print('...')
                time.sleep(30)
```
